```python
import os
SUBJECT_list="/home/a1/Desktop/sciprt/mdd_list6.txt"
BASE_DIR="/home/a1/Desktop/Projects/forkx/HC_nii"

with open(SUBJECT_list, "r") as file:
    subjects_ids = file.read().splitlines()
    
for subject_id in subjects_ids:
    INPUT_IMAGE=os.path.join(BASE_DIR,f"sub-{subject_id}","anat", f"sub-{subject_id}_T1w.nii.gz")

    if not os.path.isfile(INPUT_IMAGE):
        print(f"Input image not found: {INPUT_IMAGE}")
        continue

    recon_all_cmd = f"recon-all -s {subject_id} -i {INPUT_IMAGE} -all -sd {BASE_DIR}"

    print(f"runding reconall ")
    os.system(recon_all_cmd)
```
