```python
import subprocess
import os                                                                                                                                             

def convert_dicom_to_nifti(input_dir, output_dir):
    # Ensure the output directory exists
    os.makedirs(output_dir, exist_ok=True)

    try:
        # Run the dcm2niix command
        result = subprocess.run(
            ['dcm2niix', '','y', '-z', 'y','-o', output_dir, '-f', filename_format, input_dir],
            check=True,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            universal_newlines=True
        )
        
        # Print the output
        print("Standard Output:")
        print(result.stdout)
        
        if result.stderr:
            print("Standard Error:")
            print(result.stderr)
        
        print("Conversion completed successfully.")
        return True
    
    except subprocess.CalledProcessError as e:
        print(f"An error occurred during conversion: {e}")
        print(f"Command output: {e.output}")
        return False

# Usage
subject_dir = source_dir
target_dir = target_dir

for subfoler_name in os.listdir(subject_dir):
    subject_dir_path = os.path.join(subject_dir, subfoler_name)
    if os.path.isdir(subject_dir_path):
        input_directory = [f.path for f in os.scandir(subject_dir_path) if f.is_dir() and 'bold' in f.name]
        if input_directory:  # Check if the list is not empty
            input_directory = input_directory[0]  # Take the first matching directory
            output_directory = os.path.join(target_dir, subfoler_name,'func')
            filename_format = f'sub-{subfoler_name}_func'
            success = convert_dicom_to_nifti(input_directory, output_directory)
            if success:
                print(f"Conversion was successful for {subfoler_name}.")
            else:
                print(f"Conversion failed for {subfoler_name}.")
        else:
            print(f"No 'diff' directory found in {subfoler_name}.")
```

```python
import os
import json

def update_json_files(root_dir, task_name):
    for subdir, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith('_task-restingsz_bold.json'):
                json_file_path = os.path.join(subdir, file)
                
                # 打印正在处理的JSON文件
                print(f'Processing JSON: {json_file_path}')
                
                # 加载或创建 JSON 文件
                if os.path.exists(json_file_path):
                    with open(json_file_path, 'r') as f:
                        data = json.load(f)
                else:
                    data = {}

                # 更新 TaskName
                data['TaskName'] = task_name

                # 写回 JSON 文件
                with open(json_file_path, 'w') as f:
                    json.dump(data, f, indent=4)
                print(f'Updated JSON: {json_file_path}')

# 定义根目录和任务名称
root_directory = '/Volumes/shuang/depression/HC_nii'
task_name = 'restingsz'

# 调用函数
update_json_files(root_directory, task_name)

```
